```python
re_alum_fac = re.compile(r'<li><a href="(.+)">(.+)</a> (.+)</li>')
re_alum_visit = re.compile(r'<li><a href="(.+)">(.+)</a> (.+)</li>')
re_alum_visit_no_link = re.compile(r'<li>(.+), (.+)</li>')
```

```python
alum_fac_names = []
alum_fac_bios = []
alum_fac_urls = []

for s in reversed([s.strip() for s in alum_fac_raw.split("\n")]):
    match = re_alum_fac.search(s)
    alum_fac_urls.append(match.group(1))
    alum_fac_names.append(match.group(2))
    alum_fac_bios.append(match.group(3))
```

```python
alum_phd_names = []
alum_phd_bios = []
alum_phd_urls = []

for s in reversed([s.strip() for s in alum_phd_raw.split("\n")[:-1]]):
    match = re_alum_fac.search(s)
    alum_phd_urls.append(match.group(1))
    alum_phd_names.append(match.group(2))
    alum_phd_bios.append(match.group(3))
```

```python
alum_visit_names = []
alum_visit_bios = []
alum_visit_urls = []

for s in reversed([s.strip() for s in alum_visit_raw.split("\n")]):
    match = re_alum_visit.search(s)
    if match is not None:
        alum_visit_urls.append(match.group(1))
        alum_visit_names.append(match.group(2))
        alum_visit_bios.append(match.group(3))
    else:
        match = re_alum_visit_no_link.search(s)
        alum_visit_urls.append(None)
        alum_visit_names.append(match.group(1))
        alum_visit_bios.append(match.group(2))
```

```python
count = 0
count_o = 97

def get_count():
    global count
    global count_o
    c = chr(count_o)+str(count)
    if count < 9:
        count += 1
    else:
        count_o += 1
        count = 0
    return c
```

```python
def print_alum_faculty(name, bio, web):
    r = "alum-faculty-%s:\n    display_name: %s\n    role: alum-faculty\n    bio: %s\n    webpage: %s\n" % (get_count(), name, bio, web)
    return r
```

```python
alum_faculty_strs = [print_alum_faculty(alum_fac_names[i], alum_fac_bios[i], alum_fac_urls[i]) for i in range(len(alum_fac_names))]
```

```python
for s in reversed(alum_faculty_strs):
    print(s)
```

```python
def print_alum_phd(name, bio, web):
    r = "alum-phd-%s:\n    display_name: %s\n    role: alum-phd\n    bio: %s\n    webpage: %s\n" % (get_count(), name, bio, web)
    return r
```

```python
alum_phd_strs = [print_alum_phd(alum_phd_names[i], alum_phd_bios[i], alum_phd_urls[i]) for i in range(len(alum_phd_names))]
```

```python
for s in reversed(alum_phd_strs):
    print(s)
```

```python
def print_alum_visit(name, bio, web):
    if web is not None:
        r = "alum-visit/uc-%s:\n    display_name: %s,\n    role: alum-visit/uc\n    bio: %s\n    webpage: %s\n" % (get_count(), name, bio, web)
    else:
        r = "alum-visit/uc-%s:\n    display_name: %s,\n    role: alum-visit/uc\n    bio: %s\n" % (get_count(), name, bio)
    return r
```

```python
alum_visit_strs = [print_alum_visit(alum_visit_names[i], alum_visit_bios[i], alum_visit_urls[i]) for i in range(len(alum_visit_names))]
```

```python
for s in reversed(alum_visit_strs):
    print(s)
```
